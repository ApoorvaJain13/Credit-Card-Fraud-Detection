```python
logmodel.fit(X_train,y_train)
```

```python
print(confusion_matrix(y_test, predictions))
```

# KNN

TRAIN - TEST - SPLIT

```python
data_2000 = df[:20000]
X = data_2000.drop(30, axis = 1)
y = data_2000[30]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

MODEL CREATION

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate = []

for i in range (1,40):
    
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

# Performance (KNN)

```python
plt.figure(figsize = (10,6))
plt.plot(error_rate, color = 'blue', ls = 'dashed', marker = 'o',markerfacecolor = 'red', markersize = 10)
plt.title ('Error Rate')
plt.xlabel ('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors= 2)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print (confusion_matrix(y_test, pred))
print ('\n')
print (classification_report(y_test, pred))
```

# Neural Network

TRAIN - TEST - SPLIT

```python
X = df.drop(30, axis = 1).values
y = df[30].values


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Model Creation

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(20, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(10, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(5, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation = 'sigmoid'))


model.compile (optimizer = 'sgd',loss = 'binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 25)
```

```python
model.fit(x = X_train, y = y_train, epochs = 100,validation_data= (X_test,y_test), callbacks = [early_stop], batch_size= 400)
```

# Performance (Neural Network)

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print (classification_report(y_test, predictions))
```

```python
print (confusion_matrix(y_test, predictions))
```
